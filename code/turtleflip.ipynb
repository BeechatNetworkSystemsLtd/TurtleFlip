# TurtleFlip

The `TurtleFlip` package is a small Python library implementing code from the ocrodeg NVlabs library, to bypass OCR systems.

# DATA AUGMENTATION

```python
image = imread("testdata/image.png")
imshow(image)
```

# RANDOM DISTORTIONS

```python
for i, sigma in enumerate([0.75]):
    subplot(2, 2, i+1)
    noise = ocrodeg.bounded_gaussian_noise(image.shape, sigma, 5.0)
    distorted = ocrodeg.distort_with_noise(image, noise)
    h, w = image.shape
    imshow(distorted)
```

# RULED SURFACE DISTORTIONS

```python
for i, mag in enumerate([130.0]):
    subplot(2, 2, i+1)
    noise = ocrodeg.noise_distort1d(distorted.shape, magnitude=mag)
    distorted2 = ocrodeg.distort_with_noise(distorted, noise)
    h, w = distorted2.shape
    imshow(distorted2)
    
    #Outputs distorted image as JPEG
    import scipy.misc
    imsave('output-file.jpg', distorted2)
    
```
